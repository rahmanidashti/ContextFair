```python
import pandas as pd
```

```python
top_n = 10
ds_name = "Yelp"
```

```python
results = pd.read_csv(f"results/best_results_{top_n}.csv")
```

```python
# show results dataframe
results
```

```python
ds_results_group = results.groupby("Dataset")
```

### Barchart on nDCG

```python
def barchart_df_creator(dataset="Gowalla", fusion="mul"):
    fusion_results_group = ds_results_group.get_group(dataset).groupby("Fusion")
    barchart_df = pd.DataFrame(columns=["Model", "Group", "nDCG"])
    # print(fusion_results_group.get_group(fusion))
    for eachline in fusion_results_group.get_group(fusion).itertuples():
        barchart_df = barchart_df.append({"Model": eachline.Model, "Group": "Leisure", "nDCG": eachline.leisure_ndcg}, ignore_index=True)
        barchart_df = barchart_df.append({"Model": eachline.Model, "Group": "Working", "nDCG": eachline.working_ndcg}, ignore_index=True)
        barchart_df = barchart_df.append({"Model": eachline.Model, "Group": "All", "nDCG": eachline.all_ndcg}, ignore_index=True)
    return barchart_df
```

```python
barchart_df = barchart_df_creator(dataset=ds_name)
barchart_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Set Seaborn style
sns.set_style('darkgrid')

sns.barplot(x = "Model", y = "nDCG", hue = "Group", ci = None, data = barchart_df)
plt.savefig(f"plots/results/{ds_name}_barchart_{top_n}.pdf", bbox_inches='tight')
```
